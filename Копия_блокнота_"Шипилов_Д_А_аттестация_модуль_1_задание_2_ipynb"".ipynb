```python
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/sasaradwan/1/main/telecom_churn.csv')
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
df.head() 

```

```python
print (df.shape, '- размер данных')
```

```python
print ('Названия столбцов:')
print ()
print (df.columns)
```

```python
print ('Общая информация по датафрейму:')
print ()
print (df.info())
```

```python
print ('Изменение типа колонки Churn')
print()
df ['Churn'] = df ['Churn'].astype('int64')
```

```python
print ('Использование describe:')
df.describe ()
print ()
df.describe (include=['object','bool'])
```

```python
print ('Распредеелние данных по целевой переменной - Churn:')
df['Churn'].value_counts ()
```

```python
print ('Распределение пользователей по переменной Area code:')
print ()
df['Area code'].value_counts (normalize=True)
```

```python
print ('Сортировка DataFrame по столбцу Total day charge:')
print ()
df.sort_values (by='Total day charge',ascending=False).head()
```

```python
print ('Сортировка DataFrame по группе столбцов: Churn и Total day charge:')
print ()
df.sort_values (by=['Churn','Total day charge'],ascending=[True,False]).head()
```

```python
print ('Индексация DataFrame по полю Name:')
print()
print ('Какова доля людей нелояльных пользователей в нашем датафрейме?')
print (df['Churn'].mean()*100,'%')
```

```python
print ('Логическая индексация DataFrame по одному столбцу')
print ()
print ('Каковы средние значения числовых признаков среди нелояльных пользователей?')
print ()
df[df['Churn']==1].mean()
```

```python
print ('Комбинация индексирования')
print ()
print('Cколько в среднем в течение дня разговаривают по телефону нелояльные пользователи?')
print ()
print (df[df['Churn']==1]['Total day minutes'].mean(),'минут')
print ()
print ('Какова максимальная длина международных звонков среди лояльных пользователей , не пользующихся услугой международного роуминга ?')
print (df[(df['Churn'] == 0) & (df['International plan'] == 'No')]['Total intl minutes'].max(),'минут')
```

```python
print ('Индексация по названиям столбцов, строк, порядковому номеру:')
print ()
print ('Передача значений по (id) номерам от 0 до 5 для столбцов от State до Area code:')
print ()
df.loc[0:5,'State':'Area code']
```

```python
print ('Передача значений первых пяти строк в первых трех столбцах:')
print ()
df.iloc[0:5,0:3]
```

```python
print ('Отражение первой строчки датафрейма :')
print ()
df[:1]
```

```python
print ('Отражение последней строчки датафрейма :')
print ()
df[-1:]
```

```python
print ('Применение функции к каждому столбцу:')
print ()
df.apply(np.max)
```

```python
print ('Использование "map" для замены значений в колонке:')
print ()
d={'No':False,'Yes':True}
df['International plan'] = df['International plan'].map(d)
df.head()
```

```python
print ('Использование "replace" для замены значений в колонке:')
print ()
df=df.replace({'Voice mail plan':d})
df.head()
```

```python
print ('Группирование данных в зависимости от значения признака Churn и вывод статистик по трём столбцам в каждой группе.')
print()
columns_to_show = ['Total day minutes','Total eve minutes','Total night minutes']
df.groupby (['Churn'])[columns_to_show].describe (percentiles=[])
```

```python
print ('Группирование данных в зависимости от значения признака Churn и вывод статистик по трём столбцам в каждой группе, с передачев в agg списка функций')
print()
columns_to_show = ['Total day minutes','Total eve minutes','Total night minutes']
df.groupby (['Churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
print ('Построение таблицы сопряженности методом crosstab (False)')
print()
pd.crosstab (df['Churn'],df['Voice mail plan'], normalize=False)
```

```python
print ('Построение таблицы сопряженности методом crosstab (True)')
print()
pd.crosstab (df['Churn'],df['Voice mail plan'], normalize=True)
```

```python
print ('Cреднее число дневных, вечерних и ночных звонков для разных значений Area code:')
print ()
df.pivot_table(['Total day calls','Total eve calls','Total night calls'],['Area code'], aggfunc='mean').head(10)
```

```python
print ('Подсчет общего количества звонков для всех пользователей:')
print ()
total_calls = df['Total day calls'] + df['Total eve calls'] + df['Total night calls'] + df['Total intl calls']
df.insert (loc=len(df.columns), column = 'Total calls',value=total_calls)
df.head()
```

```python
print ('Подсчет общей цены звонков для всех пользователей:')
print ()
df['Total charge']=df['Total day charge'] + df['Total eve charge'] + df['Total night charge'] + df['Total intl charge']
df.head()
```

```python
print ('Ликвидация созданных столбцов:')
print ()
df.drop (['Total charge', 'Total calls'], axis=1)
```

```python
print ('Ликвидация созданных строк:')
print ()
df.drop ([1,2], axis=0).head()
```

```python
print ('Связь оттока с International plan:')
print ()
pd.crosstab (df['Churn'], df['International plan'], margins=True)
```

```python
print ('Связь оттока с International plan - график')
print ()
pd.crosstab (df['Churn'], df['International plan']).plot(kind='bar',stacked=False)
```

```python
print ('Число обращений в сервисный центр:')
print ()
pd.crosstab (df['Churn'], df['Customer service calls'], margins=True)
```

```python
print ('Число обращений в сервисный центр - график')
print ()
pd.crosstab (df['Customer service calls'],df['Churn']).plot(kind='bar',stacked=False)
```

```python
print ('Связь оттока с Customer service calls (больше 3х):')
print ()
df['Many_service_calls']=(df['Customer service calls']>3).astype('int')
pd.crosstab (df['Many_service_calls'], df['Churn'], margins=True)
```

```python
print ('Связь оттока с Customer service calls (больше 3х) - график')
print ()
df['Many_service_calls']=(df['Customer service calls']>3).astype('int')
pd.crosstab (df['Many_service_calls'], df['Churn']).plot(kind='bar',stacked=False)
```

```python
print ('Связь оттока с International plan и Числом обращений (>3):')
print ()
pd.crosstab(df['Many_service_calls'] & df['International plan'], df['Churn'])
```

```python
print ('Связь оттока с International plan и Числом обращений (>3) - график')
print ()
pd.crosstab(df['Many_service_calls'] & df['International plan'], df['Churn']).plot(kind='bar',stacked=False)
```
